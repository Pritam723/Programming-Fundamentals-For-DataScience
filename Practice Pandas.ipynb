<h3>Filtering Data</h3>

```python
# Surely .loc[] is most useful here. Because it is labelled based indexing.
# .iloc[] is position based. .loc[] takes boolean based indexing too.
```

```python
mask = obesity_data['Height'] > 1.72
mask
```

```python
# We we can pass this mask to iloc[]
# obesity_data.loc[obesity_data['Height'] > 1.72, ["Gender","Age", "Height"]]

obesity_data.loc[mask, ["Gender","Age", "Height"]]
```

```python
# Question, find the rows for which Weight < 50 and Obesity = 'Normal_Weight'

mask = (obesity_data['Weight'] < 50) & (obesity_data['Obesity'] == 'Normal_Weight')
# print(mask)
obesity_data.loc[mask]
```

```python
# So .loc[] supports boolean indexing. The thing is, even if we do don't 
# use .loc[] here it is fine.

obesity_data[mask] # gives the same output. It is shorthand notation.
# Can write the full mask too.
```

<h3>Regular Expressions</h3>

```python
mask1 = obesity_data['Obesity'].str.contains('Normal')
mask2 = obesity_data['Obesity'].str.startswith('Fat')

obesity_data.loc[mask1]
```

<h3>Transformation of Data</h3>

```python
def increment_by_5(x):
    return x+5

obesity_data["Age"].apply(lambda x: x+5)
```

<h3>Insert, Drop, Delete, Rename Columns</h3>

```python
# To add column at particular position.
obesity_data.insert(2,'BMI',obesity_data['Weight']/obesity_data['Height']**2)

# See the values, it is working because under the hood, those 2 are
# numpy arrays on which division is applied item-wise.
```

```python
obesity_data.drop(columns = ['BMI', 'Age']) # Can give singe column name too.
# But it does not change the original df.
obesity_data.drop(columns = ['BMI', 'Age'], inplace=True)
# This will change the original df.
```

```python
# Another way to delete column is to use delete
del obesity_data['CAEC'] # Will change the original df.
```

```python
help(obesity_data.rename) # Use this to know about the function.

obesity_data.rename(columns={'BMI' : 'New BMI', 'Age' : 'New Age'})
# But it is not inplace. To do that use inplace = True
```

<h3>Merge, Join & Concatenate Dataframes</h3>

```python
df1 = pd.DataFrame({'id' : [1,2,3,4], 'name' : ['a', 'b', 'c', 'd']})
df2 = pd.DataFrame({'id' : [3,4,5,6], 'course' : ['c1', 'c2', 'c3', 'c4']})

merged_df = pd.merge(df1, df2, how = 'inner', on = 'id')
merged_df
```

```python
concat_df = pd.concat([df1, df2], axis = 0, keys = ['df1', 'df2'])
concat_df
```

<h3>Handling NULL Values</h3>

```python
concat_df['name'].fillna('No Name')
```

```python
concat_df.fillna({'name' : 'no name', 'course' : 'no course'})
```
